```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

cands = [Path('mycar/data'), Path('data'), Path('sample_data')]
found = None
for d in cands:
    if d.exists():
        imgs = list(d.rglob('*.jpg')) + list(d.rglob('*.png'))
        if imgs:
            found = imgs[0]
            break

if found is None:
    print('No image files found in', cands)
    print('Put sample images into sample_data or your data folder and re-run this cell.')
else:
    print('Showing', found)
    img = Image.open(found)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
