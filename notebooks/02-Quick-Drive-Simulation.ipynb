```python
# List scripts/ directory if present, then run a small check
import time
import math
from pathlib import Path
import matplotlib.pyplot as plt
# simulate a simple 2D trajectory with steering and throttle
T = 200
x = 0.0
y = 0.0
theta = 0.0
xs, ys = [], []
for t in range(T):
    steer = 0.3 * math.sin(2*math.pi*t/40)
    throttle = 0.5 + 0.2*math.sin(2*math.pi*t/60)
    # simple bicycle model step
    theta += steer * 0.05
    x += throttle * math.cos(theta) * 0.1
    y += throttle * math.sin(theta) * 0.1
    xs.append(x)
    ys.append(y)
# plot trajectory
plt.figure(figsize=(5,5))
plt.plot(xs, ys, '-o', markersize=2)
plt.title('Simulated drive trajectory')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
# save a quick image to notebooks for visualization demos
out = Path('sample_data')
out.mkdir(parents=True, exist_ok=True)
imgp = out / 'sim_trajectory.png'
plt.savefig(imgp.as_posix())
print('Saved trajectory image to', imgp)
```
